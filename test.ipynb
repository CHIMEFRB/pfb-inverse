```python
import pfb
import h5py
```

```python
ntime = 2**20

# Generate a white noise timestream
ts = np.random.standard_normal(ntime)

# Perform the PFB
spec_pfb = pfb.pfb(ts, 129)

# Perform the inverse
rts = pfb.inverse_pfb(spec_pfb, 4)
```

```python
# Save out PFB data
np.save('testpfb.dat', spec_pfb[:, :-1])
```

```python
# Invert PFB data with prog
```

```python
%%bash

rm rects.dat
mpirun -np 4 python pfbdriver.py testpfb.dat.npy rects.dat -f 128 -n 4 -m
```

```python
# Load in recovered data
with h5py.File('rects.dat', 'r') as f:
    rts2 = f['ts'][:]
```

```python
rts2.shape
```

```python
subplot(121)
imshow(np.log10(np.abs(rts2 - ts.reshape(-1, 256)[3:])), aspect='auto')
colorbar()
subplot(122)
imshow(np.log10(np.abs(rts - ts.reshape(-1, 256))), aspect='auto')
colorbar()
```
